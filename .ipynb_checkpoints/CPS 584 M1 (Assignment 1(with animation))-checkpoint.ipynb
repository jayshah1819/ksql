```python
import numpy as np
import matplotlib.pyplot as plt

#defining sigmoid function
def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))

#adding thresold function
def is_below_threshold(dw1, dw2, dw3, threshold=0.001):
    max_change = np.max(np.abs([dw1, dw2, dw3]))
    return max_change < threshold

# Input X1 and X2
x1 = np.array([0, 0, 1, 1, 1, 2, 2])
x2 = np.array([0, 1, 0, 1, 2, 2, 3])

# Expected output
t = np.array([0, 0, 0, 0, 1, 1, 1])

# Learning rate (Rho)
R = 0.01

# Initialize weights and bias
w1, w2, w3 = np.random.uniform(-3, 0, 3)

# Number of epochs
epochs = 10000


for step in range(epochs):
    for i in range(7):
        U = x1[i] * w1 + x2[i] * w2 + w3
        # called the sigmoid function
        y = sigmoid(U)

        # Delta values
        dw1 = R * x1[i] * (t[i] - y)
        dw2 = R * x2[i] * (t[i] - y)
        dw3 = R * (t[i] - y)

        # new weights and bias
        w1 = w1 + dw1
        w2 = w2 + dw2
        w3 = w3 + dw3

        # Use is_below_threshold to update flag
        if is_below_threshold(dw1, dw2, dw3):
            flag = False

# This will create an axis
k = np.linspace(-3.5, 3.5, num=100)

# This will be the decision boundary
decision_boundary = -(w3 + k * w1) / w2

# Data points
for i in range(7):
    if t[i] == 0:
        plt.scatter(x1[i], x2[i], marker='o', c='blue')
    else:
        plt.scatter(x1[i], x2[i], marker='o', c='red')

# The decision boundary
plt.plot(k, decision_boundary, label='Decision Boundary', linestyle='-', color='green')

# Showing the final weights
print("Final Weights:", w1, w2, w3)

# Showing the New weights and the output 
print("Output for each input pattern:")
for i in range(7):
    if (x1[i] * w1 + x2[i] * w2 + w3) > 0:
        output = 1
    else:
        output = 0
    print(f"Input: ({x1[i]}, {x2[i]}), Target Output: {t[i]}, Predicted Output: {output}")

# the ploting the x1 and x2
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Neural Network Decision Boundary')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation
from IPython.display import display, clear_output
import time

# Input data
X1 = np.array([0, 0, 1, 1, 2, 3, 3])
X2 = np.array([0, 1, 0, 1, 3, 3, 4])
X3 = np.array([0, 0, 0, 0, 3, 2, 4])

# Output data
t = np.array([0, 0, 0, 0, 1, 1, 1])

# Learning rate (Rho)
R = 0.01

# Initialize weights and bias
w1 = np.random.uniform(-3, 0, 1)
w2 = np.random.uniform(-3, 0, 1)
w3 = np.random.uniform(-3, 0, 1)
w4 = np.random.uniform(-3, 0, 1)

# Number of epochs
epochs = 10000

# Create a meshgrid for 3D plotting
xx, yy = np.meshgrid(np.linspace(-1, 5, num=100), np.linspace(-1, 5, num=100))

# Set up the figure and 3D axis
fig_3d = plt.figure()
ax_3d = fig_3d.add_subplot(111, projection='3d')

# Scatter plot of data points
scatter = ax_3d.scatter(X1, X2, X3, c=t, cmap='coolwarm', marker='o')

# Plotting the initial decision boundary
zz = -(w4 + w1 * xx + w2 * yy) / w3
surface = ax_3d.plot_surface(xx, yy, zz, alpha=0.5, color='green')

# Display the final weights
print("Initial Weights:", w1, w2, w3, w4)

# Update function for animation
def update(frame):
    global w1, w2, w3, w4
    for i in range(7):
        U = X1[i] * w1 + X2[i] * w2 + X3[i] * w3 + w4
        y = 1.0 / (1 + np.exp(-U))
        
        dw1 = R * X1[i] * (t[i] - y)
        dw2 = R * X2[i] * (t[i] - y)
        dw3 = R * X3[i] * (t[i] - y)
        dw4 = R * (t[i] - y)
        
        w1 += dw1
        w2 += dw2
        w3 += dw3
        w4 += dw4
    
    zz = -(w4 + w1 * xx + w2 * yy) / w3
    surface.set_array(zz.ravel())
    
    output = [1 if (X1[i] * w1 + X2[i] * w2 + X3[i] * w3 + w4) > 0 else 0 for i in range(7)]
    scatter.set_array(np.array(output))

    print(f"Epoch: {frame}, Weights: {w1}, {w2}, {w3}, {w4}")

# Create animation
ani_3d = FuncAnimation(fig_3d, update, frames=epochs, interval=50, repeat=False)

# Display the animation
display(fig_3d)

# Stop the animation when it's done (custom condition)
try:
    for _ in range(epochs):
        time.sleep(0.05)
except KeyboardInterrupt:
    pass
```
