```python
import numpy as np
import matplotlib.pyplot as plt

# Input X1 and X2
x1 = np.array([0, 0, 1, 1, 1, 2, 2])
x2 = np.array([0, 1, 0, 1, 2, 2, 3])

# Expected output
t = np.array([0, 0, 0, 0, 1, 1, 1])

# Learning rate(Rho)
R = 0.01

# Initialize weights and bias
w1, w2, w3 = np.random.uniform(-3, 0, 3)

# Number of epoches
epochs = 10000

# Training loop
for step in range(epochs):
    for i in range(7):
        U = x1[i] * w1 + x2[i] * w2 + w3
        #sigmoid function
        y = 1.0 / (1 + np.exp(-U))

        # Delta values
        dw1 = R * x1[i] * (t[i] - y)
        dw2 = R * x2[i] * (t[i] - y)
        dw3 = R * (t[i] - y)

        # Updated weights and bias
        w1 = w1 + dw1
        w2 = w2 + dw2
        w3 = w3 + dw3

# Create an axis
k = np.linspace(-3.5, 3.5, num=100)

# this will be the decision baundary
decision_boundary = -(w3 + k * w1) / w2

#data points
for i in range(7):
    if t[i] == 0:
        plt.scatter(x1[i], x2[i], marker='o', c='blue')
    else:
        plt.scatter(x1[i], x2[i], marker='o', c='red')

#the decision boundary
plt.plot(k, decision_boundary, label='Decision Boundary', linestyle='-', color='green')

# Displaying the plot
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.title('Neural Network Decision Boundary')
plt.show()
```
